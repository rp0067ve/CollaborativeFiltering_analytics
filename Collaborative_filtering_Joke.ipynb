```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_excel("/content/jester-data-1.xls", header=None)
df
```

```python
df = df.drop(0, axis=1)
```

```python
df = df.replace(99.00, 0)
```

## user2のitem1のジョーク評価を予測していく

```python
df
```

```python
cos_list = []
for user_id in df.index:
  cos_list.append(cosine_similarity([df.loc[1, 2:]], [df.loc[user_id, 2:]]))

cos_list
```

```python
user_list = sorted(range(len(cos_list)), key=lambda i: cos_list[i], reverse=True)
user_list
```

```python

weighted_sum = 0
total_weight = 0

for user_id in user_list[1:]:
  weight = cos_list[user_id]
  review_score = df.iloc[user_id][1]
  if review_score > 0:
    weighted_sum += review_score * weight
    total_weight += weight

weighted_avg = weighted_sum / total_weight
print(weighted_avg)
```
