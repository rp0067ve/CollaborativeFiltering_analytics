```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.data", names=["user_id", "item_id", "review", "timestamp"], sep="\t")
```

```python
df = pd.pivot_table(df.drop("timestamp", axis=1), index="user_id", columns="item_id", values="review")
```

```python
df = df.fillna(0)
```

```python
df
```

```python
cos_list = []
for user_id in df.index:
  cos_list.append(cosine_similarity([df.loc[2, 2:]], [df.loc[user_id, 2:]]))
cos_list
```

```python
user_list = sorted(range(len(cos_list)), key=lambda i: cos_list[i], reverse=True)
```

```python
user_list
```

```python
weighted_sum = 0
total_weight = 0
for user_id in user_list[1:]:
  weight = cos_list[user_id]
  review_score = df.iloc[user_id][1]
  if review_score > 0:
    weighted_sum += review_score * weight
    total_weight += weight

print(weighted_sum)
print(total_weight)
weighted_avg = weighted_sum / total_weight
print(weighted_avg)
```

```python
df = pd.read_csv("https://files.grouplens.org/datasets/movielens/ml-100k/u.data", names=["user_id", "item_id", "review", "timestamp"], sep="\t")
df = pd.pivot_table(df.drop("timestamp", axis=1), index="user_id", columns="item_id", values="review")

df
```

```python
# ãƒ¦ãƒ¼ã‚¶ãƒ¼2ã®ã‚¢ã‚¤ãƒ†ãƒ 1ã®è©•ä¾¡ã‚’ä¸€æ™‚çš„ã«ä¿å­˜ã—ã¦Nanã§åŸ‹ã‚ã‚‹
user2_item1_review = df.loc[2, 1]
df.loc[2, 1] = np.nan

# å¹³å‡å€¤åŸ‹ã‚
df = df.apply(lambda x:x.fillna(x.mean()), axis=1)
```
